```python
import pandas as pd
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = 'Documents/Anexo 2 - Dataset Vehicle/car details v4.csv'
data = pd.read_csv(file_path)
```

```python
data_info = data.info()
data_head = data.head()
```

```python
data_info, data_head
```

```python
plt.figure(figsize=(14, 10))
for i, column in enumerate(['Make', 'Model', 'Price', 'Year','Color','Location'], 1):
    plt.subplot(3, 2, i)
    sns.histplot(data[column], kde=True)
    plt.title(f'Distribution of {column}')

plt.tight_layout()
plt.show()
```
